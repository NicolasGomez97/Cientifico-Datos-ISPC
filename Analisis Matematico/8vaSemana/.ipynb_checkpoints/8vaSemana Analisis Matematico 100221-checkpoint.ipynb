Supongamos que en $x=x_1$ la derivada de la funcin $y=f(x)$ se reduce a cero, es decir, $f'(x)=0$. Admitamos, adems que existe la segunda derivada, $f''(x)$, y es continua sobre cierta vecindad del punto $x_1$. Para este caso es vlido el siguiente teorema:

> **Teorema 1.** Si $f'(x_1)=0$, entonces en $x=x_1$ la funcin tiene mximo cuando $f''(x_1)<0$, y, un mnimo cuando $f''(x_1)>0$.

De acuerdo al teorema anterior, para caracterizar los puntos crticos de una funcin $f(x)$ es necesario utilizar el criterio de la segunda derivada. Entendiendo que los puntos crticos se obtiene de resolver la ecuacin $f'(x)=0$.

En Python es posible realizar clculo simblico mediante la librera SymPy.

---

**Ejemplo 1.** *Calcular y caracterizar los puntos crticos de la funcin $f(x) = x^3 - x$*

Primero, y como siempre, importamos los mdulos necesarios.

## Clculo de mximos y mnimos

 

Para encontrar los extremos relativos o locales de una funcin f(x),  realizaremos lo siguiente:

1) Hallar la primera derivada f'(x) y obtener sus races.

 

2) Hallar la segunda derivada f''(x), y calcular los valores que toman los ceros de la primer derivada en f''(x), luego, determinar si es un mximo o mnimo de acuerdo a la condicin, recordando que si:

 

f''(x) > 0 Tenemos un mnimo

f''(x) < 0 Tenemos un mximo

 

3) Calcular la segunda coordenada de los extremos relativos en la funcin f(x).


### Pregunta: Que pasa si f''(x) = 0?

```python
from sympy import *
from sympy.plotting import plot
from sympy.abc import x
import sympy as sp
init_printing()
```

Definimos la funcin con la cual operaremos:

```python
f = x**3 - x
```

Acto seguido, calculamos la primera y segunda derivada de $f(x)$ mediante la funcin `diff`:

```python
df = diff(f, x) # 1era. derivada
d2f = diff(f,x,2) # Segunda derivada
df, d2f
```

Los puntos crticos los calculamos resolviendo la ecuacin $f'(x)=0$

```python
pc = solve(Eq(df, 0))
pc
```

Para determinar si se trata de un mnimo o mximo, utilizamos el criterio de la segunda derivada, sustituyendo los puntos crticos en $f''(x)$, es decir:

```python
d2f.subs(x,pc[0]) # Primer punto crtico
```

```python
d2f.subs(x,pc[1]) # Segundo punto crtico
```

Con esto determinamos, acorde al teorema 1, que el primer punto crtico $\left(\frac{-\sqrt{3}}{3}\right)$ es un mnimo y el segundo $\left(\frac{\sqrt{3}}{3}\right)$ un mximo. Podemos comprobarlo trazando la grfica correspondiente:

```python
plot(f, (x, -1.2, 1.2))
```

Podemos *automatizar* un poco este proceso definiendo una funcin que realice el procedimiento descrito.

```python
def maxminf(f):
    """ Calcula los mximos y mnimos de una funcin f(x) """
    df = diff(f, x) # 1era. derivada
    d2f = diff(f, x, 2) # 2da. derivada
    pcs = solve(Eq(df,0)) # puntos crticos
    for p in pcs:
        if d2f.subs(x,p)>0: 
            tipo="Min"
        elif d2f.subs(x,p)<0: 
            tipo="Max"
        else: 
            tipo="Indefinido"
        print("x = %f (%s)"%(p,tipo))
```

```python
maxminf(x**3 - x)
```

La funcin anterior puede arrojar errores cuando se tienen races complejas, lo cual podra considerarse mediante una estructura try-except o bien verificando si el punto crtico es un valor real.

## Derivadas parciales aplicadas a machine learning

### Error Cuadrtico Medio

https://www.iartificial.net/error-cuadratico-medio-para-regresion/

El Error Cuadrtico Medio es el criterio de evaluacin ms usado para problemas de regresin. Se usa sobre todo cuando usamos aprendizaje automtico supervisado. Para cada dato histrico podremos indicar el resultado correcto. 

Vamos a ver como se calcula.

$error $ $ cuadratico=(real  estimado)2$

$ y_{pred} = ax_{i}^2 + bx_i + c $

$R = \displaystyle\sum_{i=1}^n{(y_i-(a*x_i^2+b*x_i+c))^2}$

### Qu es el Gradiente?

https://www.iartificial.net/gradiente-descendiente-para-aprendizaje-automatico/

Creo que la mejor forma de explicar lo que es el gradiente, es diciendo que es una generalizacin de la derivada. Seguramente ya sabis lo que es una derivada. Aqu tenis un pequeo recordatorio.

En matemticas, la derivada de una funcin mide la rapidez con la que cambia el valor de dicha funcin matemtica, segn cambie el valor de su variable independiente. La derivada de una funcin es un concepto local, es decir, se calcula como el lmite de la rapidez de cambio media de la funcin en cierto intervalo, cuando el intervalo considerado para la variable independiente se torna cada vez ms pequeo. Por ello se habla del valor de la derivada de una funcin en un punto dado.

$\displaystyle\frac{R}{a}=2\sum_{i=1}^n{((yi(ax2i+bxi+c))x2i)}$

```python
from sklearn.metrics import mean_squared_error
y_true = [3, -0.5, 3, 9]
y_pred = [2.5, 0.0, 2, 8]
mse = mean_squared_error(y_true, y_pred)
mse
# resultado: 0.375
```

```python
import numpy as np
y_true = np.array([3, -0.5, 3, 9])
y_pred = np.array([2.5, 0.0, 2, 8])
mse = (np.square(y_true - y_pred)).mean()
mse
# resultado: 0.375
```
